```python
import os
class escala:
    def __init__(self):
        
        self.escalaCromatica = [['sol', 0], ['sol#', 0], ['la', 0], ['la#', 0], ['si', 0],
            ['do', 1], ['do#', 1], ['re', 1], ['re#', 1], ['mi', 1], ['fa', 1], ['fa#', 1], ['sol', 1], ['sol#', 1],  ['la', 1], ['la#', 1], ['si', 1],
            ['do', 2], ['do#', 2], ['re', 2], ['re#', 2], ['mi', 2], ['fa', 2], ['fa#', 2], ['sol', 2], ['sol#', 2],  ['la', 2], ['la#', 2], ['si', 2]]
        self.dir_fitxers = os.path.join(os.getcwd(),"fitxers")
        
        self.escalaCromaticaUnaOctava = ['do', 'do#', 're', 're#', 'mi', 'fa', 'fa#', 'sol', 'sol#','la','la#', 'si']
    
    
    def getNotesEscala(self, nota,  patro):
        if len(patro) != 7:
            raise TypeError("Introdueix un patro de 7 intervals")
        
        #Reordena la llisa per començar amb la nota que toca
        llista_repositori = []
        llista_bona = []
        notaTrobada = False
        for i in self.escalaCromaticaUnaOctava:
            if i == nota:
                notaTrobada = True
                
            if notaTrobada == True:
                llista_bona.append(i)
            else:
                llista_repositori.append(i)
        
        llista_bona = llista_bona + llista_repositori
        
        # Amb l'ordre correcte aplico el patro passat per parametre
        index_comencar = 0
        lst_final = []
        
        for i in patro:
            lst_final.append(llista_bona[index_comencar])
            try:
                if i == "S":
                    index_comencar += 1
                if i == "T":
                    index_comencar += 2
            except IndexError:
                print("Error a l'index.")
                
        return lst_final
                    

    
    def ordenarEscalaCromatica(self ,nota, octavica=False, bidireccional=False):
        try:
            posicioNota = self.escalaCromatica.index(nota)
            if octavica == True:
                escala_octava = self.escalaCromatica[posicioNota:posicioNota+13]
                escala_octava_rev = escala_octava[::-1]
                if bidireccional == True:
                    return escala_octava + escala_octava_rev
                return self.escalaCromatica[posicioNota:posicioNota+13]
            return self.escalaCromatica[posicioNota:]
        except:
            return self.escalaCromatica
    def ferArpegi(self, patro, anariTornar = False):
        llistaArpegi = [patro[0] , patro[2] , patro[4], patro[-1]]
        if anariTornar == False:
            return llistaArpegi
        if anariTornar == True:
            return llistaArpegi + llistaArpegi[::-1]

    def ferEscala(self, notaa, patro, unaOctava = True, anariTornar=False, afegirArpegi=False):
  
        # Un patro es T, T, S, T, T, S, T
        posicionsArpegi = [0, 2, 4, 7, 9,11, 14]
        #llistaArpegi = self.ferArpegi(patro, anariTornar)
        escalaPartir = self.ordenarEscalaCromatica(notaa, False, False)
        maxIndex = len(escalaPartir)
        dic_patrons = {"T":2, "S":1}
        llistaIndexEscalaBuscada = []
        llistaArpegi = []
        pos = 0
        if unaOctava == True:
            for i in range(len(patro)):
                llistaIndexEscalaBuscada.append(pos)
                if i in posicionsArpegi:
                    llistaArpegi.append(pos)
                pos += dic_patrons[patro[i]]
        if unaOctava == False:
            patro = patro*2
            for i in range(len(patro)):
                if pos < maxIndex:
                    llistaIndexEscalaBuscada.append(pos)
                    if i in posicionsArpegi:
                        llistaArpegi.append(pos)
                pos += dic_patrons[patro[i]]
            
        #+ ["Separador", "Gadgeds"]
        llistaNotesFinal = [escalaPartir[i] for i in llistaIndexEscalaBuscada]
        llistaNotesFinalArpegi = [escalaPartir[i] for i in llistaArpegi]
        if anariTornar == False:
            if afegirArpegi == True:
                print("  --+", llistaNotesFinalArpegi)
                llistaNotesFinal.append(["Separador", "Gadgeds"])
                return llistaNotesFinal  + llistaNotesFinalArpegi
            return llistaNotesFinal
        if anariTornar == True:
            if afegirArpegi == True:
                
                llistaTronada = llistaNotesFinal[::-1]
                llistaTronada.append(["Separador", "Gadgeds"])
                return llistaNotesFinal + llistaTronada   + llistaNotesFinalArpegi + llistaNotesFinalArpegi[::-1]
            return llistaNotesFinal + llistaNotesFinal[::-1]
        
    def getEscalaPerMastil(self, nota, patro):
        llista_retorn = self.ferEscala(nota, patro, True, False, False)
        llista_adaptada = [f"{i[0]}{i[1]}" for i in llista_retorn]
        return llista_adaptada
            
        



    def retornarPartituraeEscala(self, nota, patro, unaOctava=True, anariTornar=False,  afegirArpegi=False):
        return partitura().crearPartitura(self.ferEscala(nota,patro, unaOctava, anariTornar, afegirArpegi))
    
    def guardarPartitura(self, nom, nota, patro, unaOctava=True, anariTornar=False, afegirArpegi=False):
        arxiu = open(os.path.join(self.dir_fitxers ,f"partitura_{nom}.txt"), "w")
        
        arxiu.write(partitura().retornarPartituraeEscala(nota, patro, unaOctava, anariTornar, afegirArpegi))
        
        arxiu.close()
        
class patronsEscala:
        #Major
        major = ["T", "T", "S", "T", "T", "T", "S"]
        
        #Menors
        menorNatural = ["T", "S", "T", "T", "S", "T", "T"]
        menorHarmonica = ["T", "S", "T", "T", "S", "S", "T"]
        menorMelodicaAsc = ["T", "S", "T", "T", "T", "T", "S"]
        menorMelodicaDesc = ["T", "T", "S", "T", "T", "S", "T"]
        
        #Modals
        jonica = ["T", "T", "S", "T", "T", "T", "S"]
        dorica = ["T", "S", "T", "T", "T", "S", "T"]
        frigia = ["S", "T", "T", "T", "S", "T", "T"]
        lidia = ["T", "T", "T", "S", "T", "T", "S"]
        mixolidia = ["T", "T", "S", "T", "T", "S", "T"]
        eolica = ["T", "S", "T", "T", "S", "T", "T"]
        locria = ["S", "T", "T", "S", "T", "T", "T"]
        frigia = ["S", "T", "T", "T", "S", "T", "T"]
    

class partitura:
    def __init__(self):

        self.escalaCromatica = [['sol#', 0], ['sol', 0], ['la', 0], ['la#', 0], ['si', 0],
        ['do', 1], ['do#', 1], ['re', 1], ['re#', 1], ['mi', 1], ['fa', 1], ['fa#', 1], ['sol', 1], ['sol#', 1],  ['la', 1], ['la#', 1], ['si', 1],
        ['do', 2], ['do#', 2], ['re', 2], ['re#', 2], ['mi', 2], ['fa', 2], ['fa#', 2], ['sol', 2], ['sol#', 2],  ['la', 2], ['la#', 2], ['si', 2]]
        

    def combinar_llistes(self, *llistas):
        combinades = []
        combinadesStr = ""
        for elementos in zip(*llistas):
            combinades.append("".join(elementos))
            combinadesStr += "".join(elementos)
        return combinadesStr

    
    def ferLlistaNotes(self, args):
        final = ['  \n', ' - \n', '   \n', ' - \n', '   \n', '---\n', '   \n', '---\n', '   \n', '---\n',
                    '   \n', '---\n', '   \n', '---\n', '   \n', '   \n', '   \n'][::-1]
        
        llistaNotes = []
        for i in args:
            llista = nota(i[0], i[1]).getLlista()
            llistaNotes.append(llista)

        llistaNotes.append(final)
    
        return llistaNotes
   
    def crearPartitura(self, llistaNotes):
        return self.combinar_llistes(*self.ferLlistaNotes(llistaNotes))
    
    def retornarPartituraeEscala(self, nota, patro, unaOctava=True, anariTornar=False,  afegirArpegi=False):
        return self.crearPartitura(escala().ferEscala( nota, patro, unaOctava, anariTornar,  afegirArpegi))
    

    
        
        
        




#print(prova.retornarPartituraeEscala(["do", 1], ["T", "T", "S", "T", "T", "T", "S"]))


# print(escala().guardarPartitura("ProvaSol", ["sol", 0], ["T", "T", "S", "T", "T", "T", "S"], False, True, True))
#escalaImprimir = escala().retornarPartituraeEscala(["do", 1], patronsEscala.mixolidia , False, True, True)


#print(escalaImprimir)
#escala().aplicarPatroEscala("do", patronsEscala().menorHarmonica)
#escala().getEscalaPerMastil("mi", patronsEscala().dorica)

```
